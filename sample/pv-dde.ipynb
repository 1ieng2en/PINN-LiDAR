推导声学波动方程，以质点速度（particle velocity）为目标，首先需要写出相关的基本方程。

连续性方程（质量守恒）： 这是基于质量守恒原则的方程，描述了密度随时间和空间的变化。

$[ \frac{\partial \rho}{\partial t} + \nabla \cdot (\rho \mathbf{v}) = 0 ]$

其中，(\rho) 为密度，(\mathbf{v}) 为质点速度。

动量守恒方程： 该方程基于牛顿第二定律（动量守恒），描述了力对质点速度的影响。

$[ \rho \frac{\partial \mathbf{v}}{\partial t} + \rho (\mathbf{v} \cdot \nabla) \mathbf{v} = -\nabla p + \mathbf{f} ]$

其中，(p) 为声压，(\mathbf{f}) 为体积力密度。

状态方程（可以表述为声速与密度、声压的关系）：

在本文中，简化为线性无耗声学条件下的一般线性方程：

$[ p = c^2 (\rho - \rho_0) ]$

其中，(c) 为声速，(\rho_0) 为静态密度（未扰动流体的密度）。

接下来，我们将这些方程结合，推导以质点速度为目标的波动方程：

首先，假设流体的小振幅波，因此可以线性化连续性方程：

$[ \frac{\partial \rho'}{\partial t} + \rho_0 \nabla \cdot \mathbf{v} = 0 ]$

接着是线性化的动量方程（忽略非线性项和体积力）：

$[ \rho_0 \frac{\partial \mathbf{v}}{\partial t} = -\nabla p ]$

将状态方程代入上述动量方程：

$[ \rho_0 \frac{\partial \mathbf{v}}{\partial t} = -\nabla (c^2 \rho') ]$

结合线性化连续性方程得到：

$[ \frac{\partial \rho'}{\partial t} = -\rho_0 \nabla \cdot \mathbf{v} ]$

代入到动量方程中：

$[ \rho_0 \frac{\partial \mathbf{v}}{\partial t} = -c^2 \nabla (\rho_0 \nabla \cdot \mathbf{v}) ]$

从而得到质点速度(\mathbf{v})的波动方程：

$[ \nabla^2 \mathbf{v} = \frac{1}{c^2} \frac{\partial^2 \mathbf{v}}{\partial t^2} ]$

这样，我们就推导出了以质点速度为目标的波动方程。这个过程展示了如何将基本的连续性和动量守恒原理应用于声学问题，得到质点速度的波动方程。

在声学中，状态空间模型可以用来描述声波传播的动态特性。针对质点速度，我们可以通过以下步骤构建状态空间模型。

1. 定义状态变量
假设声压 (p) 和质点速度 (\mathbf{v}) 为两个状态变量。可以定义状态向量 (\mathbf{x}(t)) 为：

$[ \mathbf{x}(t) = \begin{bmatrix} \mathbf{v} \ p \ \end{bmatrix} ]$

2. 状态方程
基于之前推导的线性化方程，我们可以得到以下两个方程：

连续性方程

$[ \frac{\partial \rho'}{\partial t} = -\rho_0 \nabla \cdot \mathbf{v} ]$

动量方程

$[ \rho_0 \frac{\partial \mathbf{v}}{\partial t} = -\nabla p ]$

结合状态方程，(\rho' = \frac{p}{c^2})：

$[ \frac{\partial p}{\partial t} = -c^2 \rho_0 \nabla \cdot \mathbf{v} ]$

将这两个方程转化为状态空间形式。将偏导数换成导数简化形式：

$[ \frac{d}{dt} \begin{bmatrix} \mathbf{v} \ p \ \end{bmatrix}
\begin{bmatrix} 0 & -\frac{1}{\rho_0} \nabla \ -c^2 \rho_0 \nabla \cdot & 0 \ \end{bmatrix} \begin{bmatrix} \mathbf{v} \ p \ \end{bmatrix} ]$

3. 输出方程
假设输出方程仅为质点速度，可以表示为：

$[ \mathbf{y}(t) = \begin{bmatrix} I & 0 \end{bmatrix} \begin{bmatrix} \mathbf{v} \ p \ \end{bmatrix} ]$

4. 完整的状态空间模型
结合状态方程和输出方程，系统的状态空间模型可以表示为：

$[ \frac{d}{dt} \mathbf{x}(t) = A \mathbf{x}(t) + B \mathbf{u}(t) ]$

$[ \mathbf{y}(t) = C \mathbf{x}(t) + D \mathbf{u}(t) ]$

其中，矩阵 (A), (B), (C), (D) 为：

$[ A = \begin{bmatrix} 0 & -\frac{1}{\rho_0} \nabla \ -c^2 \rho_0 \nabla \cdot & 0 \ \end{bmatrix}, \quad B = \begin{bmatrix} 0 \ 0 \ \end{bmatrix}, \quad C = \begin{bmatrix} I & 0 \ \end{bmatrix}, \quad D = \begin{bmatrix} 0 \ \end{bmatrix} ]$

当考虑一个点声源作为外力，并从质点速度的角度推导波动方程的理论解时，我们需要将外力项添加到动量方程中，并进行相应的处理。

### 方程修正：

1. **动量守恒方程（包含外力项）**：
   
   加入点声源的外力\( \mathbf{f}(\mathbf{x}, t) \)：

   $[
   \rho_0 \frac{\partial \mathbf{v}}{\partial t} = -\nabla p + \mathbf{f}(\mathbf{x}, t)
   ]$

   其中，\(\mathbf{f}(\mathbf{x}, t)\)是点声源的表示，可以记作强度为\(F\)的脉冲函数\(\delta(\mathbf{x})\delta(t)\)，即：

   $[
   \mathbf{f}(\mathbf{x}, t) = F \delta(\mathbf{x}) \delta(t)
   ]$

2. **与连续性方程结合**：

   再次写出线性化的连续性方程：

   $[
   \frac{\partial \rho'}{\partial t} = -\rho_0 \nabla \cdot \mathbf{v}
   ]$

3. **结合动量和连续性方程**：

   利用状态方程（$[p = c^2 \rho']$），结合继续推导：

   $[
   \rho_0 \frac{\partial \mathbf{v}}{\partial t} = -c^2 \nabla ( \rho_0 \nabla \cdot \mathbf{v} ) + F \delta(\mathbf{x}) \delta(t)
   ]$

   代入连续性方程得到波动方程：

   $[
   \nabla^2 \mathbf{v} - \frac{1}{c^2} \frac{\partial^2 \mathbf{v}}{\partial t^2} = \frac{F}{\rho_0 c^2} \nabla \delta(\mathbf{x})\delta(t)
   ]$

### 理论解的推导：

为了求解上述偏微分方程，我们可以利用傅立叶变换或者格林函数。此处采用格林函数的策略：

1. **格林函数求解**：
   
   二次三维波动方程的格林函数\( G(\mathbf{x}, t) \)满足：

   $[
   \nabla^2 G - \frac{1}{c^2} \frac{\partial^2 G}{\partial t^2} = \delta(\mathbf{x})\delta(t)
   ]$

   其解在三维空间中为：

   $[
   G(\mathbf{x}, t) = \frac{\delta(t - |\mathbf{x}|/c)}{4\pi |\mathbf{x}|}
   ]$

2. **应用到质点速度的波动方程**：

   通过卷积操作，我们可以得到质点速度的解：

   $[
   \mathbf{v}(\mathbf{x}, t) = \frac{F}{\rho_0 c^2} \nabla G(\mathbf{x}, t)
   ]$

   这意味着质点速度场是点声源引起的格林函数的梯度的结果。

3. **最终解的形式**：

   通过计算，可得质点速度的解为：

   $[
   \mathbf{v}(\mathbf{x}, t) = \frac{F}{4\pi \rho_0 c^2 |\mathbf{x}|} \hat{\mathbf{x}} \delta(t - |\mathbf{x}|/c)
   ]$

   其中$\hat{\mathbf{x}}$是从声源指向观察点的方向单位向量。

这个解展示了在含有点声源的情况下质点速度场的理论解法，以反映声波传播过程中的瞬态行为，并提供了对声场分布的理解。
